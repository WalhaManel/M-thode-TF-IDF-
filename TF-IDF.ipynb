# tokenization & droping stop-words & Lemmatisation

```python

def tokenization(data):
    tk = WhitespaceTokenizer()
    data_words=[tk.tokenize(row) for row in data ]
    return data_words
def clean(doc):
    crow=[re.sub(r"[]"+ string.punctuation + r"]", " ",r) for r in doc]
    mysp_fr=stopwords.words('french')
    mysp_ang=stopwords.words('english')
    mysp=mysp_fr+mysp_ang
    #©@~&£$*%§°#
    crow=[re.sub(r"[0-9©@~&£$*%§°#’]"," ",w) for w in crow]
    crow=[re.sub(r'\s+', '',  w) for w in crow]
    crow=[w for w in crow if (w not in mysp) & (w != '') & (w != " ")]
    lemmatizer = WordNetLemmatizer()
    cdoc=[lemmatizer.lemmatize(w, pos='v') for w in crow]
    return cdoc
data=tokenization(data)
data=[clean(doc) for doc in data]

```

# 1- Term Frequency (TF)

```python
#Placer tous les mots qui apparaissent dans le texte dans un "Sac"
bagOfWords=[]
for l in data:
    bagOfWords+=l
'''
Function qui prend en paramétre un documents et 
retourner un dictionnaire contient les mots de bagOfWords et
ses occurances dans le documents donnée en paramétre 
'''
def numOfWords(doc):
    numOfWords={w:0 for w in doc}
    for w in bagOfWords:
        if numOfWords.get(w) != None :
            numOfWords[w]+=1
        else :
            numOfWords[w]=0
    return numOfWords
```

```python
'''
Fonction qui compte les Fréquences des Termes 
'''
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
for doc in data :
    tfD = computeTF(numOfWords(doc), doc)
    print(tfD)
```

# Inverse Data Frequency (IDF)

```python
'''
Fonction qui compte La fréquence inverse des mots dans les documents 
'''
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict

idfs = computeIDF([numOfWords(doc) for doc in data])
print(idfs)
```
